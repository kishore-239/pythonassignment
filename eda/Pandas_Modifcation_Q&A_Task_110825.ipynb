## IPL Data Analysis Questions & Answers

### 1. How many unique batsmen are there?

```python
uniqueBatsmen = df['batsman'].nunique()
print("The no.of Unique batsmen are :", uniqueBatsmen)
```

### 2. What is the total number of extra runs in IPL history?

```python
extra_runs_total = df['extra_runs'].sum()
print("the total number of extra runs in IPL history is:", extra_runs_total)
```

### 3. What is the total number of runs scored in IPL history?

```python
runs_total = df['total_runs'].sum()
print("the total number of runs scored in IPL history is :", runs_total)
```

### 4. What are the different types of dismissals, their counts, and percentages?

```python
dismissals_types = df['dismissal_kind'].value_counts()
dismissals_percentages = round((dismissals_types / dismissals_types.sum()) * 100, 2)

print("the different types of dismissals, their counts, and percentages are :")
print("=="*40)
print()

for i in dismissals_types.index:
    print(f"{i}: {dismissals_types[i]} ({dismissals_percentages[i]}%)")


```

### 5. Show records where total runs in a ball are more than 6.

```python
runs_more_than_6 = df[df['total_runs'] > 6]
print("Records where total runs in a ball are more than 6 :", len(runs_more_than_6))
```

### 6. Show only batsman, batting_team, bowling_team, and bowler where runs > 6 in a ball.

```python
columns_new = runs_more_than_6[['batsman', 'batting_team', 'bowling_team', 'bowler']]
columns_new.head()
```

### 7. How many total runs were made by BB McCullum against Royal Challengers Bangalore?

```python
runs_by_mccullum = df[(df['batsman'] == 'BB McCullum') & 
                   (df['bowling_team'] == 'Royal Challengers Bangalore')]['batsman_runs'].sum()
print("BB McCullum runs against Royal Challengers Bangalore were :", runs_by_mccullum)
```

### 8. How many unique bowlers have bowled for Mumbai Indians?

```python
Mumbai_Indians_bowlers = df[df['bowling_team'] == 'Mumbai Indians']['bowler'].nunique()
print("The no.of unique bowlers have bowled for Mumbai Indians was :", Mumbai_Indians_bowlers)
```

### 9. What are the different types of dismissals for Sunrisers Hyderabad as the batting team?

```python
types_of_dismissals = df[df['batting_team'] == 'Sunrisers Hyderabad']['dismissal_kind'].dropna().unique()
print("The different types of dismissals for Sunrisers Hyderabad as the batting team are :\n\n", types_of_dismissals)
```

### 10. How many runs did Rohit Sharma score against Sunrisers Hyderabad while batting for Mumbai Indians?

```python
Rohit_Sharma_runs = df[(df['batsman'] == 'RG Sharma') & 
                (df['batting_team'] == 'Mumbai Indians') & 
                (df['bowling_team'] == 'Sunrisers Hyderabad')]['batsman_runs'].sum()
print("Rohit Sharma score against Sunrisers Hyderabad while batting for Mumbai Indians :", Rohit_Sharma_runs)
```

### 11. Which batsman played the most balls?

```python
most_balls_faced = df['batsman'].value_counts()
most_balls_batsman = most_balls_faced.idxmax()
print("The batsman who played the most balls is :", most_balls_batsman)
```

### 12. Which bowler bowled the most valid deliveries (excluding wides and no-balls)?

```python
bowled_valid_deliveries = df[(df['wide_runs'] == 0) & (df['noball_runs'] == 0)]
bowler_counts = bowled_valid_deliveries['bowler'].value_counts()
most_valid_bowler = bowler_counts.idxmax()
print("The bowler who bowled the most valid deliveries (excluding wides and no-balls) is :", most_valid_bowler)
```

### 13. How many extra runs were conceded by Delhi Capitals against Mumbai Indians?

```python
extra_runs_dc_against_mi = df[(df['bowling_team'] == 'Delhi Capitals') & 
                      (df['batting_team'] == 'Mumbai Indians')]['extra_runs'].sum()
print("The extra runs conceded by Delhi Capitals against Mumbai Indians were :", extra_runs_dc_against_mi,"runs")
```
