#### Q3) [1512] Number of Good Pairs

```python
from collections import Counter
from typing import List

def numIdenticalPairs(nums: List[int]) -> int:
    ans = 0
    for c in Counter(nums).values():
        ans += c * (c - 1) // 2
    return ans
```

```python
numIdenticalPairs([1,2,3,2,3,3,1])
```

#### Q4)  [1365] How Many Numbers Are Smaller Than the Current Number

```python
from typing import List

class Solution(object):
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        cnt = [0] * 101
        for x in nums:
            cnt[x] += 1

        # prefix sum: cnt[v] = how many numbers <= v
        for v in range(1, 101):
            cnt[v] += cnt[v - 1]

        return [0 if x == 0 else cnt[x - 1] for x in nums]
```

```python
s = Solution()
print(s.smallerNumbersThanCurrent([8, 1, 2, 2, 3]))
```

#### Q5) [1281] Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        prod = 1
        add = 0
        while n>0:
            digit = n%10
            prod *= digit
            add += digit

            n //= 10
        return prod - add
```

```python
s = Solution()
print(s.subtractProductAndSum(234))
```

#### Q6) [1486] XOR Operation in an Array

```python
from typing import List

class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        ans = 0
        for i in range(n):
            ans ^= start + 2 * i
        return ans
```

```python
s = Solution()
print(s.xorOperation(5, 0))
```
