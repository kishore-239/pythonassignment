## Accessing the dataframe content - Indexing and Slicing Data

### Using `Dataframe[ ]`

```python
# extract single column as Pandas Data Series
# df_name['col_name'] 

df['customerID']
type(df['customerID'])

# extract single column as Pandas Data Frame
# df_name[['col_name']] 
df[['customerID']]
type(df[['customerID']])
```

```python
# extract multiple columns
# df_name[['col1', 'col2'...'coln']] 

df[['customerID', 'InternetService', 'PaymentMethod']]
```

```python
#  extract one row - nth row; where n is the row index num
# df_name[n:n+1]  
df[44:45]
```

```python
# extract multiple rows from m to n row index with step num i
# df_name[m:n:i]  
df[1:5]
df[1:5:2]
df[1:20:5]
df[20:1:-5]
df.iloc[:, [1, 2]]
```

```python
# extract mutiple rows & cols
# df_name[m:n:i][['col1', 'col2'...'coln']] 

df[1:44:9][['customerID', 'InternetService', 'PaymentMethod']]
```

### Indexing Pandas Data Series


```python
df.InternetService
df.InternetService[2] # single value from Series; 1D array indexing 
df.InternetService[2:9]
df.InternetService[2:44:9]
```

```python
# get the top 3 brands by count 
df.customerID.value_counts() 
df.customerID.value_counts()[:3]

# get the bottom 3 brands by count 
df.customerID.value_counts()[-3:]
```

### Using Dataframe.loc[ ]

```python
df.loc[::,::] # Entire DF 
```

```python
# get only specific rows 
# get the rows from Index no. 1 to 3 
df.loc[1:3]
df.loc[1:3,]
df.loc[1:3,::]
```

```python
# get only specific columns 
# get the columns - make, fuel_type,aspiration,num_of_doors,body_style
df.loc[:,'customerID']
df.loc[:,['customerID']]
df.loc[:,['customerID', 'PaymentMethod']]
df.loc[:,'customerID':'InternetService']
df.loc[:,'customerID':'InternetService':3]
```

```python
# apply custom row indx numbers 

# Assign new vaues to DF index  - Row1,Row2,Row3,....,Row7043 
df.index = [f"Row{i}" for i in range(1,7044)]
df.head()
```

```python
df.loc["Row1"] # Extracting Single row using ".loc": Output is Series 
df.loc[["Row1", "Row9", "Row44"]] # Extracting multiple rows using ".loc"
df.loc["Row9":"Row18":2] # Extracting multiple rows in dequence using ".loc"
```

```python
# reset the DF index back to 0,1,2,....,7043 
df.index = range(7043)
df.head() 
df.index
```

### Using Dataframe.iloc[ ]

```python
# complete DF 
df.iloc[::,::]
```
