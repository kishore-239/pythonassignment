## 5. What is the average tenure of customers with OnlineBackup?


```python
a = df[df['OnlineBackup'] == 'Yes']['Tenure'].mean()
print(a)
```

```python
b = df.groupby('Contract')['TotalCharges'].mean()
print(b)
```

## 8. What is the churn rate by gender?


```python
d = df.groupby('Gender')['Churn'].apply(lambda x: (x == 'Yes').mean() * 100)
print(d)
```

## 13. Compare the average TotalCharges for customers with and without TechSupport.

## 14. Among non-churned customers, which services (OnlineSecurity, OnlineBackup, etc.) are most common?


```python
a_non_churned_customers = df[df['Churn'] == 'No']
b_service_columns = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
c = a_non_churned_customers[b_service_columns].apply(lambda x: (x == 'Yes').sum()).sort_values(ascending=False)
```
