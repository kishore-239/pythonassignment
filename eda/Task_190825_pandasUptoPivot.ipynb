### Q:
Create a Series from a NumPy array and perform mean, median, mode, variance, and standard deviation on the array.


````python

a = np.array([10, 20, 20, 30, 40, 50])

# Convert NumPy array to Pandas Series
b = pd.Series(a)
print(b)
````

````python

# Mean
c = b.mean()
print(c)

# Median
d = b.median()
print(d)

# Mode
e = b.mode()[0]  # mode() returns a Series
print(e)

# Variance
f = b.var()
print(f)

# Standard Deviation
g = b.std()
print(g)
````

### Q:
1. Perform an **inner join** on `df1` and `df2` using the common column `ID` and display the result.  
2. Perform a **left join** on `left_df` and `right_df` using the common column `ID` and display the result.  
3. Perform a **cross join** on `df7` and `df8` and display the result.


````python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

inner_join_df = pd.merge(df1, df2, on='ID', how='inner')
inner_join_df
````

````python
left_df = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
right_df = pd.DataFrame({'ID': [2, 3, 4], 'Age': [25, 30, 22]})

left_join_df = pd.merge(left_df, right_df, on='ID', how='left')
left_join_df
````

````python
df7 = pd.DataFrame({'Color': ['Red', 'Green', 'Blue']})
df8 = pd.DataFrame({'Size': ['Small', 'Medium', 'Large']})

cross_join_df = df7.merge(df8, how='cross')
cross_join_df
````

### Q:
1. Create project stages based on progress: 'Early', 'Mid', 'Late'.  
2. Find the average progress for each department.  
3. Create a pivot table to display the total progress for each month and department.


````python
project_data = {
    'ProjectID': [101, 102, 103, 104, 105],
    'Department': ['HR', 'IT', 'HR', 'IT', 'HR'],
    'Month': ['Jan', 'Feb', 'Jan', 'Feb', 'Jan'],
    'Progress': [80, 60, 90, 75, 85]
}

df = pd.DataFrame(project_data)
df
````

````python
def stage(progress):
    if progress < 70:
        return 'Early'
    elif 70 <= progress < 85:
        return 'Mid'
    else:
        return 'Late'

df['Stage'] = df['Progress'].apply(stage)
df
````

````python
a = df.groupby('Department')[['Progress']].mean()
a
````

````python
b = pd.pivot_table(df, values='Progress', index='Month', columns='Department', aggfunc='sum')
b
````
